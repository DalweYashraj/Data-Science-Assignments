```python
import pandas as pd
import numpy as np
```

```python
book_df=pd.read_csv("book.csv")
```

```python
book_df.head()
```

```python
#number of unique users in the dataset
len(book_df.User_ID.unique())
```

```python
len(book_df.Book_Title.unique())
```

```python
user_book_df=book_df.pivot_table(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_df.User_ID.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_df.User_ID.unique()
user_sim_df.columns = book_df.User_ID.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```
